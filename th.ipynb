# IMPORTATION DE LIBS

```python
import pandas as pd
import seaborn as se
import matplotlib.pyplot as plt
```

## Importation de base de données


```python
df1 = se.load_dataset('titanic')
df1.head()
```

## Manipulation des données

```python
se.countplot(x="survived",data=df1)



```

```python
se.countplot(x="survived",hue="pclass",data=df1)
```

```python
se.countplot(x="survived",hue="sex",data=df1)
```

```python
fig = plt.figure(figsize=(20,8))
survived = df1['survived']
sex = df1['sex']
df_male = survived[sex == 'male'].value_counts().sort_index()
df_female = survived[sex == 'female'].value_counts().sort_index()

ax1 = fig.add_subplot(121)
df_male.plot(kind='barh',label='Male', alpha=0.55)
df_female.plot(kind='barh', color='#FA2379',label='Female', alpha=0.55)
plt.title("Who Survived? with respect to Gender, (raw value counts) "); plt.legend(loc='best')
ax1.set_ylim(-1, 2) 
```
